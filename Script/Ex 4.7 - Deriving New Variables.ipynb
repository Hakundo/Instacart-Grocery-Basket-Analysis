#### Sort your products into price ranges groups by creating `flags`

Apply the following criteria: 

- If the item’s price is lower than or equal to $5, it will be labeled a “low-range product.”

- If the item’s price is above USD 5 but lower than or equal to USD 15, it will be labeled a “mid-range product.”

- If the item’s price is above USD 15, it will be labeled a “high-range product.”

#### Create a subset to avoid any potential issues that can lead to memory issues or processing power because of the larga dataframe

```python
df = ords_prods_merge[:1000000] # Everything from beginning up to the number
```

```python
df.shape
```

#### Define a function to sort data by categories

```python
df.head()
```

```python
def price_label(row): # creating new function - name of the function - argument (row)

  if row['prices'] <= 5:                              # If the item’s price is lower than or equal to $5
    return 'Low-range product'                        # it will be labeled a “low-range product
  elif (row['prices'] > 5) and (row['prices'] <= 15): # If the item’s price is above $5 but lower than or equal to $15
    return 'Mid-range product'                        # it will be labeled a “mid-range product.”
  elif row['prices'] > 15:                            # If the item’s price is above $15
    return 'High range'                               # it will be labeled a “high-range product.”
  else: return 'Not enough data'                      # If any of the above can't be found then return "Not enough data"
```

#### Apply functions


**`df['price_range']`** : cretates new column called **price_range** designated as location for price_label

**`df.apply(price_label, axis=1)`** : tells Python to **apply** the **price_label** function on **axis=1**

**axis=1** stands for "**row**" so this code essentially tells Python to apply the function to all **`rows`** within the dataframe.

**axis=0** Would tell Python to apply the function to all **`columns`** within the dataframe

```python


df['price_range'] = df.apply(price_label, axis=1)
```

### 2) If-Statements with the loc() Function

Using **loc()**, you can apply the conditional logic of an if-statement to a function without explicitly creating an **if-els**e construct.

#### Create conditions as the previous method

```python
# 1)

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# 2)

df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# 3) 

 
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False) # value.counts(): counts number o products within each label 
```

#### 1)

- **if** = `df.loc[df['prices'] > 15,`

- **then** = `'price_range_loc'] = 'High-range product'`


#### 2)

- **if** = `df.loc[(df['prices'] <= 15) `**&**` (df['prices'] > 5),` These two conditions are combined by the **&** sign in the middle of the “if” half.
- **then** = `'price_range_loc'] = 'Mid-range product'`


#### 3)

- **if** = `df.loc[df['prices'] <= 5,`

- **then** = `'price_range_loc'] = 'Low-range product'`

### Why loc() over User-defined-function?


1. Using **loc()** won’t result in a warning message

2. **loc()** method runs much faster; the **loc()** function applies the conditional filters before searching through the dataframe, while your **user-defined function** searches through the entire dataframe and then determines where to set the filters (remember **axis = 1**?).

### 3) If-Statements with For-Loops

**`For-loops`**, are **loops** for running the same block of **code multiple times**. Perform the same function on multiple elements, for instance, by running through an entire dataframe and performing a function on each row within that dataframe.

Example:

Print the statement **“My age is X”** for ages 30 to 45 (and substituting the age in question for “X” for each one). Without a **for-loop**, manually print statements. 

```python
print("My age is 30")
```

```python
print("My age is 31")
```

```python
print("My age is 32")
```

```python
print("My age is 33")
```

#### With for-loop

Yu can print everything with a single command

```python
for x in range(30,45):
    print("My age is %d" % (x)) 
```

#### Function breakdown

The **loop** cycles for every value **(x)** within the range of 30 and 45. Each time it cycles, it executes the code inside the loop, which, in this case, prints `“My age is”` followed by the current value of x

**"My age is %d"** is a string that includes a placeholder **%d**, which will be replaced by the value of the variable **x**.
**`% (x)`** is used to format the string, replacing **%d** with the current value of **x**.

### Create a new column in the ords_prods_merge dataframe that summarizes thow busy each day of the week is

```python
# 1) : First we need to know on which day most orders take place by printing the frequency of the “orders_day_of_the_week” column

ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

Brief: 

0 = Saturday
1 = Sunday
2 = Monday
3 = Tuesday
4 = Wednesday
5 - Thursday
6 - Friday

Result breakdown:

**0** has the highest frequency, meaning **Saturday** is the busiest day and **4**, **Wednesday**, is the slowest day of orders

We need the information to create a new column, **busiest day** that will contain: "Busiest day", "Least busy", and "Regurlarly busy". 

```python
result = []                                         # Create an empty list " result"

for value in ords_prods_merge["orders_day_of_week"]:# For each value within the column "orders of week" we want something
  if value == 0:                                    # if the value of the row is equal to 0
    result.append("Busiest day")                    # then "Busiest day" is appended to the blank "result" list
  elif value == 4:                                  # if the value is equal to 4
    result.append("Least busy")                     # then " Least busy" is appended to the "result" list
  else:                                             # if is neither of these conditions are met
    result.append("Regularly busy")                 # then 'Regularly busy'

# "value" acts as placeholder, it could be called anything, could be also an X. 
# Here represents every possible age that could exist between 30 and 45. 
# Here represents "orders_day_of_week"
```

```python
# After running loop, print gthe result list

result
```

The long list procuded does not do anything by itself. We need to combine it with **`ords_prods_merge`** datafame to can be used effectively. 

To combine it, we need to create a new column within **`ords_prods_merge`** and **`result`**

```python
ords_prods_merge['businest_day'] = result
```

```python
# Print the frequency of the new column and cross-check it with the frequency of "orders_day_of_week"

ords_prods_merge['businest_day'].value_counts(dropna = False)
```

With that, you’ve successfully created a summary column for the busiest day of the week!

# Task 4:

When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled **“Most orders”**,  **“Average orders”**,  and **“Fewest orders”** Create a new column containing these labels called **“busiest_period_of_day”**.

1) First we need to know on which hour of the day most orders take place by printing the frequency of the **"order_hour_of_day”** column


```python

ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

Answer: 

- **"Most orders"** = From 9:00 to 16:00

- **"Fewest orders"** = From 23:00 to 6:00

- **"Average orders"** = From 17:00 to 22:00 and from 7:00 to 8:00


```python
ords_prods_merge.shape
```

2) Create the list with the categories 

```python
result_2 = []

for value in ords_prods_merge['order_hour_of_day']:
    if (value >= 9 and value <= 16):                # a) From 9:00 to 16:00
        result_2.append("Most orders")              # Then "Most orders"
    elif (value >= 23) or (value <= 6):             # b) From 23:00 to 6:00
        result_2.append("Fewest Orders")            # then "fewest orders"
    elif (value >= 17 and value <= 22) or (value >= 7 and value <= 8): # From 17:00 to 22:00 and from 7:00 to 8:00
        result_2.append("Average orders")                              # Then "Average Orders"
    else:
        result_2.append()
        
#
```

#### Comments:

**a)** In this case, you have two conditions (value >= 9 and value <= 16) joined together with the logical operator "and"

**b)** In this case, you have two sets of conditions: (value >= 17 and value <= 22) and (value >= 7 and value <= 8). Both conditions are combined by the **or**


3) Create Column

```python
ords_prods_merge['businest_period_of_day'] = result_2
```
