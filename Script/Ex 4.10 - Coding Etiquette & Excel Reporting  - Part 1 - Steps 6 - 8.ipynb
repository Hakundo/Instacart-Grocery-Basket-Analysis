# Task 6

## 1) Age Group Distribution

```python
# Create a bar chart
age_group_distribution_bar = df_high_act_customer_dept['age_group'].value_counts().sort_index().plot.bar(color="tab:blue")

# Add labels and a legend
plt.xlabel("Age Group")
plt.ylabel("Frequency")
plt.title("Age Groups and Frequency")
plt.legend()

# Display the chart
plt.show()
```

```python
# Export

age_group_distribution_bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_group_distribution.png'))
```

## 2) Income Group Distribution

```python
# Create a bar chart
income_group_distribution_bar = df_high_act_customer_dept['income_group'].value_counts().sort_values(ascending = True).plot.barh(color="green")

# Add labels and a legend
plt.xlabel("Frequency")
plt.ylabel("Income Group")
plt.title("Income Groups and Frequency")
plt.legend()

# Display the chart
plt.show()
```

```python
# Export

income_group_distribution_bar.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'income_group_distribution.png'))
```

## 3) Parents Group Distribution

```python
# Data
data = df_high_act_customer_dept['dependants_group'].value_counts(dropna=False)
labels = data.index
sizes = data.values

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title('Dependents Group Distribution')

# Export
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'dependants_group_distribution-pie.png'))

# Display the chart
plt.show()
```

## 6) Age Group

```python
agg_age = df_high_act_customer_dept.groupby('age_group').agg({
    'order_id': ['count'],
    'prices': ['max', 'mean', 'min'],
    'days_since_last_order': ['max','mean', 'min'],
}).reset_index()

agg_age
```
