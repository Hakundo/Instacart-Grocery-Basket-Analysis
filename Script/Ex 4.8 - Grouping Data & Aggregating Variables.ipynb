## Task 4

Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the **`transform()`** and **`loc()`** functions.

#### Create a loyalty flag using loc

 Create column "max_order"         =  dataframe grouped by "user_id" column  =    get the max of values(order_number) using transform function and **`np.max`** argument

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(100)
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check the frequency of your new “loyalty_flag” column:

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

## Task 5

The marketing team at Instacart wants to know whether there’s a difference between the spending habits of the three types of customers you identified. Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). What you’re trying to determine is whether the prices of products purchased by loyal customers differ from those purchased by regular or new customers.

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'max', 'min']})
```

```python
# Group the data by the 'loyalty_flag' column
loyalty_groups = ords_prods_merge.groupby('loyalty_flag')

# Calculate basic statistics for product prices within each group
loyal_customer_stats = loyalty_groups.get_group('Loyal customer')['prices'].describe()
regular_customer_stats = loyalty_groups.get_group('Regular customer')['prices'].describe()
new_customer_stats = loyalty_groups.get_group('New customer')['prices'].describe()

# Print the statistics for each loyalty category
print("Statistics for Loyal Customers:")
print(loyal_customer_stats)

print("\nStatistics for Regular Customers:")
print(regular_customer_stats)

print("\nStatistics for New Customers:")
print(new_customer_stats)
```

## Task 6

The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the `prices` of the items people are buying. `Create a `**`spending flag`** for each **user** based on the **average price** across all their orders using the following criteria:

- If the **mean** of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
- If the **mean** of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”


```python
# Create new column "average price" using transform()

ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)

#  Create column "average_price" 
#  dataframe grouped by "user_id" column  
#  get the mean of values(prices) using transform function and **`np.mean`** argument
```

```python
ords_prods_merge.head()
```

```python
# Rounding averages to make the number smaller

ords_prods_merge['average_price'] = ords_prods_merge['average_price'].round(2)
```

#### Create the flag "Low spender"

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spending_flag'] = 'Low_spender'

#  If the average(mean) of the prices of the products by user is lower than 10 
# then(,) create a column "spending_flag where the customer will be labeled a `“Low_spender.”`
```

#### Create the flat "High spender"

```python
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spending_flag'] = 'High_spender'

#  If the average(mean) of the prices of the products by user is higher or equal to 10 
# then(,) create a column "spending_flag where the customer will be labeled a `“High_spender.”`
```

```python
ords_prods_merge.head()
```

After executing the code to set the flags, print the frequency of your new **“spending_flag”** column using the **`value_counts()`** function:

```python
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

## Task 7

In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine **frequent** versus **non-frequent** customers. Create an **order frequency flag** that marks the regularity of a **user’s ordering behavior** according to the **median** in the **“days_since_prior_order”** column. The criteria for the flag should be as follows:

- If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
- If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
- If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
# Create new column "median_days" using transform()

ords_prods_merge['median_days_since_last_order'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)

#  Create column "median_days" 
#  dataframe grouped by "user_id" column  
#  get the median of values(days_since_last_order) using transform function and **`np.median`** argument
```

```python
ords_prods_merge.head()
```

#### Create the flag "Non-frequent customer'

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'

#  If the median of the "days_prior_last_order" by user is higher than 20 
# then(,) create a column "order_frequency_flag" where the customer will be labeled a “Non-frequent customer.”
```

#### Create the flag "Regular customer"

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_since_last_order'] > 10) & (ords_prods_merge['median_days_since_last_order'] <=20), 'order_frequency_flag'] = 'Regular customer'

#  If the median of the "days_prior_last_order" by user is higher than 10 
# and lower or equal to 20 
# then(,) create a column "order_frequency_flag" where the customer will be labeled a “Regular customer.”
```

#### Create the flag "Frequent customer"

```python
ords_prods_merge.loc[ords_prods_merge['median_days_since_last_order'] <= 10, 'order_frequency_flag'] = 'Frequent customer'

#  If the median of the "days_prior_last_order" by user is lower or equal to 10 
# then(,) create a column "order_frequency_flag" where the customer will be labeled a “Frequent customer"
```

```python
ords_prods_merge.head()
```

After executing the code to set the flags, print the frequency of your new **“order_frequency_flag”** column using the **`value_counts()`** function:

```python
ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
